## 2. `Distribution Plots`

- used for univariate analysis
- used to find out the distribution
- Range of the observation

- is the data bimodal?
- Are there outliers?

Plots under distribution plot and this are axes level functions

- histplot
- kdeplot
- rugplot     

figure level function

- displot

### histplot

>* plotting uninvariate histogram

```python
tips.head()
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')

# this is figure level function
```

```python
sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
```

>* Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. Discrete bins are automatically set for categorical variables
>* **Note:** Technically this type of plot will not be called as histogram, it will be called as countplot

```python
sns.displot(data=tips, x='day', kind='hist')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

>* element -->> step

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.displot(data=titanic, x='age', kind='hist', element='step')
```

```python
sns.displot(data=titanic, x='age', kind='hist', element='step', hue='sex')
```

>* facet using col and row
>* this will not work on the histplot function because histplot is a axes level function

```python
sns.displot(data=titanic, x='age', kind='hist', element='step', col='sex')
```

### kdeplot - Kernel Density Estimation
>* Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

```python
sns.kdeplot(data=tips,x='total_bill')
# by using the axis level function
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
# by using the figure level function
```

>* hue -> fill

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

>* we can not increase the size of the plot in figure level function as we do in the matplotlib library, here we will add two parameter to increase the size of the plot.
>* one is height parameter and another one is aspect

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,
            height=10,aspect=2)
```

### rugplot
>* Plot marginal distributions by drawing ticks along the x and y axes.
>* This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

### Bivariate histogram
>* A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color

```python
sns.histplot(data=tips, x='total_bill', y='tip')
# by using axes level function
```

```python
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
# by using the figure level function
```

### Bivariate Kdeplot
>* a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

## 3. `Matrix Plot`

- Heatmap
- Clustermap

### Heatmap
>* Plot rectangular data as a color-encoded matrix

```python
gap.head()
```

>* we want to plot the year wise life expectancy of countries

```python
temp_df = gap.pivot(index="country", columns="year", values="lifeExp")
temp_df
```

```python
# heatmap is an axes level function
# here we can re size the graph by giving figure values same as matplotlib

plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

>* annot

```python
gap[gap['continent'] == 'Europe']
```

```python
gap[gap['continent'] == 'Europe'].pivot(
    index="country", columns="year", values="lifeExp")
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(
    index="country", columns="year", values="lifeExp")
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True)

# to give the annotation of the values we wil use the annot parameter
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True, linewidth=0.7, cmap='winter')

# we are using the linewidth parameter to improve readability of heatmap
# cmap is for changin the colour gradient of the graph
```
