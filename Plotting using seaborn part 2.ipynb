```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## 4. `Categorical Plots`

### 4.1 Categorical Scatter Plot

- Stripplot
- Swarmplot

#### `4.1.1 Stripplot`

```python
tips.head()
```

```python
sns.stripplot(data=tips, x='day', y='total_bill')
```

>* jitter parameter

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
# stripplot function is an axes level function
```

```python
# plotting the same graph with figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

```python
#  we can change the value of the jitter 
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', jitter=0.3)
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', jitter=0.3,
           hue='sex')
```

#### `4.1.2 Swarmplot`
>* it will be effective when the dataset is small in size

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
# axes level function
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
# plotting the swarmplot with figure level function
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='sex')
```

### 4.2 Categorical Distribution Plots

>* distribution plots will give the description of a single variable

- Boxplot
- Violinplot

#### `4.2.1 Boxplot`
>* A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). 
>* It can tell you about your outliers and what their values are. 
>* Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.
<img src='boxplot.png' height='450' width='650' />

```python
sns.boxplot(data=tips,x='sex',y='total_bill')
# plotting the graph using axes level function
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill', kind='box')
# plotting the above graph using the figure level function
```

```python
sns.boxplot(data=tips,x='day',y='total_bill', hue='sex')
```

```python
# single boxplot -> provide only numerical col
sns.boxplot(data=tips,y='total_bill')
```

#### `4.2.2 Violinplot`
>* Violinplot = (Boxplot + KDEplot)

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
# using the hue parameter
```

### 4.3 Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

#### `4.3.1 Barplot`

```python
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker')
```

```python
import numpy as np

sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

```python
import numpy as np

sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.median)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',ci=None)
# removing the error bar by providing the ci=none
```

>* When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

#### `4.3.1 Pointplot`

```python
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill')
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill', hue='sex')
```

#### `4.3.1 Countplot`
>* A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
sns.countplot(data=tips,x='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

## 5. `Regression Plots`

- regplot
- lmplot

In the simplest invocation, above both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
tips.head()
```

### regplot
>* it is axes level function
>* hue parameter is not available in this function

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

### Implot
>* hue parameter is available here
>* it is a figure level function

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

### residual plot

```python
sns.residplot(data=tips, x='total_bill', y="tip")
```

## A second way to plot Facet plots -> FacetGrid

>* **normal way of creating FacetGrid**

```python
#  figure level -> relplot -> displot -> catplot -> lmplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

>* **another way of creating FacetGrid**

```python
g = sns.FacetGrid(data=tips, col='day', row='time')
g
```

```python
g = sns.FacetGrid(data=tips, col='day', row='time')
g.map(sns.violinplot,'sex','total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='day', row='time')
g.map(sns.boxplot,'sex','total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='day', row='time', hue='smoker')
g.map(sns.scatterplot,'tip','total_bill')
g.add_legend()
```

## 6. `Multiplots`

### 6.1 Plotting Pairwise Relationship (PairGrid Vs Pairplot)

>* **Pairplot**

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

>* **PairGrid**

```python
g = sns.PairGrid(data=iris)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# we can give particular variables to get the required graph
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### 6.2 JointGrid Vs Jointplot

>* **Jointplot**

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='kde',hue='sex')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

>* **JointGrid**

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.scatterplot,sns.histplot)
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
